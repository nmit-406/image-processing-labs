```python
import random
import numpy as np
```

```python
def computeLoss(gre,toefl,urating,var,y):
    loss=0
    for j in range(1,501):
        ypred=(gre[j]/340)*var[0]+(toefl[j]/120)*var[1]+(urating[j]/5)*var[2]+var[3]
        loss+=(ypred-y[j])**2
    return loss/500
```

```python
def computeGrad(gre,toefl,urating,var,y,k):
    grad=0
    for i in range(1,501):
        ypred=(gre[i]/340)*var[0]+(toefl[i]/120)*var[1]+(urating[i]/5)*var[2]+var[3]
        if k!=3:
            grad+=(ypred-y[i])*var[k]
        else:
            grad+=(ypred-y[i])
    return grad/500
```

```python
gre=[]
toefl=[]
urating=[]
y=[]
data = np.genfromtxt('https://raw.githubusercontent.com/nmit-406/image-processing-labs/master/forward-propa%2C10-22/Admission_Predict_Ver1.1.csv',delimiter=',')
for i in data:
    gre.append(i[1])
    toefl.append(i[2])
    urating.append(i[3])
    y.append(i[8])
```

```python
var=[]
var.append(random.random())
var.append(random.random())
var.append(random.random())
var.append(random.random())
```

```python
epoch=1000
learning_rate=0.001
for i in range(0,epoch):
    print("appemt "+str(i)+": loss: "+str(computeLoss(gre,toefl,urating,var,y)))
    print("appemt "+str(i)+": a: "+str(var[0]))
    print("appemt "+str(i)+": b: "+str(var[1]))
    print("appemt "+str(i)+": c: "+str(var[2]))
    print("appemt "+str(i)+": d: "+str(var[3]))
    var[0]=var[0]-learning_rate*computeGrad(gre,toefl,urating,var,y,0)
    var[1]=var[1]-learning_rate*computeGrad(gre,toefl,urating,var,y,1)
    var[2]=var[2]-learning_rate*computeGrad(gre,toefl,urating,var,y,2)
    var[3]=var[3]-learning_rate*computeGrad(gre,toefl,urating,var,y,3)
    
```
