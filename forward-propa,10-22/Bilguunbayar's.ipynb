```python
import random
import numpy as np
```

```python
def computeLoss(gre,toefl,urating,var,y):
    loss=0
    for j in range(1,501):
        ypred=((gre[j]/340)*var[0]+(toefl[j]/120)*var[1]+(urating[j]/5)*var[2]+var[3])*var[4]
        ypred+=((gre[j]/340)*var[5]+(toefl[j]/120)*var[6]+(urating[j]/5)*var[7]+var[8])*var[9]
        ypred+=var[10]
        loss+=(ypred-y[j])**2
    return loss/500
```

```python
def computeGrad(gre,toefl,urating,var,y,k):
    grad=0
    for j in range(1,501):
        ypred=((gre[j]/340)*var[0]+(toefl[j]/120)*var[1]+(urating[j]/5)*var[2]+var[3])*var[4]
        ypred+=((gre[j]/340)*var[5]+(toefl[j]/120)*var[6]+(urating[j]/5)*var[7]+var[8])*var[9]
        ypred+=var[10]
        if k>=0 and k<=2:
            grad+=(ypred-y[j])*var[k]*var[4]
        elif k>=5 and k<=7:
            grad+=(ypred-y[j])*var[k]*var[9]
        elif k==3:
            grad+=(ypred-y[j])*var[4]
        elif k==8:
            grad+=(ypred-y[j])*var[9]
        elif k==4:
            grad+=(ypred-y[j])*((gre[j]/340)*var[0]+(toefl[j]/120)*var[1]+(urating[j]/5)*var[2]+var[3])
        elif k==9:
            grad+=(ypred-y[j])*((gre[j]/340)*var[5]+(toefl[j]/120)*var[6]+(urating[j]/5)*var[7]+var[8])
        else:
            grad+=(ypred-y[j])
    return grad/500
```

```python
gre=[]
toefl=[]
urating=[]
y=[]
data = np.genfromtxt('https://raw.githubusercontent.com/nmit-406/image-processing-labs/master/forward-propa%2C10-22/Admission_Predict_Ver1.1.csv',delimiter=',')
for i in data:
    gre.append(i[1])
    toefl.append(i[2])
    urating.append(i[3])
    y.append(i[8])
```
