```python
def computeLoss(X1,X2,X3,Y):
    loss=0
    n = len(X1)
    for i in range(n):
        loss+=((a*X1[i]+b*X2[i]+c*X3[i]+d)*r+ (e*X1[i]+f*X2[i]+g*X3[i]+h)*j+k- Y[i])**2
    
        return loss
    
```

```python
def computeGrad(X1,X2,X3,Y,i,checkk):
    togtmol=2*((a*X1[i]+b*X2[i]+c*X3[i]+d)*r+ (e*X1[i]+f*X2[i]+g*X3[i]+h)*j+k- Y[i])
    grad=0
    if(checkk=="a"):
        grad=togtmol*X1[i]*r
    if(checkk=="b"):
        grad=togtmol*X2[i]*r
    if(checkk=="c"):
        grad=togtmol*X3[i]*r
    if(checkk=="d"):
        grad=togtmol*r
    if(checkk=="r"):
        grad=togtmol*(a*X1[i]+b*X2[i]+c*X3[i]+d)
    if(checkk=="e"):
        grad=togtmol*X1[i]*j
    if(checkk=="f"):
        grad=togtmol*X2[i]*j
    if(checkk=="g"):
        grad=togtmol*X3[i]*j
    if(checkk=="h"):
        grad=togtmol*j
    if(checkk=="j"):
        grad=togtmol*(e*X1[i]+f*X2[i]+g*X3[i]+h)
    if(checkk=="k"):
        grad=togtmol
        
    return grad
```
