```python
#2dkernel
#edge-detect
ker2=[[0,-1,0],[-1,4,-1],[0,-1,0]]
```

```python
def pading2d(big,kernel):
    tmpWidth=len(kernel[0])
    tmpHeight=len(kernel)
    z = np.zeros((len(big), tmpWidth-1), dtype=big.dtype)
    # aragshaa zalgah
    big = np.concatenate((big,z), axis=1)
    # dooshoo zalgah
    for i in range(tmpHeight-1):
        zz= np.zeros((1,len(big[0])), dtype=big.dtype)
        big = np.append(big,zz, axis=0)
    return big
```

```python
r,g,b =data[:,:,0],data[:,:,1], data[:,:,2]
pr,pg,pb=pading2d(r,ker2),pading2d(g,ker2),pading2d(b,ker2)
height = len(ker2)
width = len(ker2[0])
detectedr= np.copy(r)
detectedg= np.copy(g)
detectedb= np.copy(b)
# r.setflags(write=1)
for i in range(len(pr)-height+1):
    for j in range(len(pr[0])-width+1):
        detectedr[i][j]=\
            np.sum(pr[i:i+height, j:j+width]*ker2)
        detectedg[i][j]=\
            np.sum(pg[i:i+height, j:j+width]*ker2)
        detectedb[i][j]=\
            np.sum(pb[i:i+height, j:j+width]*ker2)
```
